```python
fig = plotly.figure_factory.create_candlestick(df.open, df.high, df.low, df.close, dates=df.datetime)
fig.show()

#Espejado Vertical
df = fn.ohlc_mirror_v(df)
fig = plotly.figure_factory.create_candlestick(df.open, df.high, df.low, df.close, dates=df.datetime)
fig.show()

#Espejado Horizontal
df = fn.ohlc_mirror_h(df)
fig = plotly.figure_factory.create_candlestick(df.open, df.high, df.low, df.close, dates=df.datetime)
fig.show()
```
